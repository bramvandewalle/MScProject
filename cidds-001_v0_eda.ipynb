# Load, clean, save

```python
week1 = pd.read_csv('data/cidds-001/original/traffic/OpenStack/CIDDS-001-internal-week1.csv')
week2 = pd.read_csv('data/cidds-001/original/traffic/OpenStack/CIDDS-001-internal-week2.csv')
week3 = pd.read_csv('data/cidds-001/original/traffic/OpenStack/CIDDS-001-internal-week3.csv')
week4 = pd.read_csv('data/cidds-001/original/traffic/OpenStack/CIDDS-001-internal-week4.csv')
```
